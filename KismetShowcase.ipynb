```python
def render_relationships(name,relationships,show=set()):
    
    from graphviz import Digraph

    dot = Digraph(name)
    for pair in relationships:
        for relation in relationships[pair]:
            if relation not in show:
                continue
            dot.edge(pair[0],pair[1],_attributes=[('label',relation)])

    dot.render(name,format='png')
    

    from IPython.display import IFrame, Image
    display(Image(filename=f'{name}.png'))
#render_relationships('all',world.relationships)
render_relationships('bar',world.get_relationships('bar'),show=set(['spied_on','dating','cheating','hates','dislikes']))
```
